## Visualize data

choropleth map

```python
immigration_agg = immigration_hdi.groupby(['year', 'region']).sum().reset_index()
```

```python
fig = px.choropleth(
    data_frame=immigration_agg,
    locations='region',
    locationmode='country names',
    color='value',
    hover_name='region',
    animation_frame='year',
    projection='natural earth',
    basemap_visible = False, #change color of other countries in the map to white
    color_continuous_scale='YlOrBr',#set color scheme of bar
    range_color=(0, 600000)) # fix scale of color bar throughout animation


fig.update_layout(
    title='European Data by Country and Year',
    geo=dict(
        showframe=False,
        showcoastlines=False,
        projection_type='winkel tripel',
    fitbounds='locations'),
        geo_scope='world'
    )

#remove border lines
fig.update_traces(marker_line_width=0)

#show ocean as blue color
fig.update_geos(showocean=True,oceancolor="LightBlue")
```

bar chart

```python
immigration_agg_hdi = immigration_hdi.groupby(['year', 'region','hdi']).sum().reset_index()
immigration_agg_hdi
```

```python
fig = px.bar(
    data_frame=immigration_agg_hdi,
    x='region',
    y='value',
    color='hdi',
    animation_frame='year',
    labels={'Value': 'Number of Immigrants'},
    title='Number of Immigrants by Country of Arrival and HDI of origin country',
    barmode='group',
    range_y=[0,1000000])


fig.update_layout(
    xaxis={'title': 'Country'},
    yaxis={'title': 'Number of Immigrants'},
    legend_title='HDI of Country of Origin'
)

#get all countries in the df
unique_countries = immigration_agg_hdi['region'].unique()

# Create dropdown options
dropdown_options = []
for country in unique_countries:
    dropdown_options.append({
        'label': country,
        'method': 'update',
        'args': [{'visible': [country == c for c in immigration_agg_hdi['region']]}]
    })


#add dropdown button
fig.update_layout(
    xaxis={'title': 'Country', 'categoryorder': 'array', 'categoryarray': immigration_agg_hdi['region'].unique()},
    yaxis={'title': 'Number of Immigrants'},
    legend_title='Type of Immigrant',
    updatemenus=[
        {
            'buttons': dropdown_options,
            'direction': 'down',
            'showactive': True,
            'x': 0.1,
            'xanchor': 'left',
            'y': 0.1,
            'yanchor': 'top',
            'type':'dropdown'
        }
    ]
)
```
